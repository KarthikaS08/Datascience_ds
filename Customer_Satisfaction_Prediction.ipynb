```python
df.drop(['Customer Type', 'Age', 'Type of Travel','Flight Distance', 'Seat comfort',
       'Departure/Arrival time convenient', 'Food and drink',
       'Gate location'] , axis = 1 , inplace = True)
```

```python
df.drop(['Inflight wifi service','Leg room service','Baggage handling','Departure Delay in Minutes',
       'Arrival Delay in Minutes'], axis = 1 , inplace = True)
```

```python
df.head()
```

```python
x = df.drop('satisfaction' , axis = 1)
```

```python
x.head()
```

```python
y = df['satisfaction']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y, test_size = 0.30 ,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 500)
```

```python
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```
